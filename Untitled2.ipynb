```python
# 1. Install the necessary libraries
!pip install -q diffusers transformers accelerate opencv-python

# 2. Import the tools
import torch
from diffusers import StableVideoDiffusionPipeline
from diffusers.utils import load_image, export_to_video

# 3. Load the Model
print("Downloading the model... this takes about a minute...")
pipe = StableVideoDiffusionPipeline.from_pretrained(
    "stabilityai/stable-video-diffusion-img2vid-xt",
    torch_dtype=torch.float16,
    variant="fp16"
)
pipe.enable_model_cpu_offload() # This line is now fixed!

# 4. Get an image (We will use a rocket image for this test)
image = load_image("https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/svd/rocket.png")
image = image.resize((1024, 576))

# 5. GENERATE VIDEO
print("Generating video... please wait...")
# We use a 'seed' so the result is the same every time you run it
frames = pipe(image, decode_chunk_size=8, generator=torch.manual_seed(42)).frames[0]

# 6. Save the video
export_to_video(frames, "my_first_ai_video.mp4", fps=7)
print("Success! Your video is saved as 'my_first_ai_video.mp4'")
```

```python
from IPython.display import HTML
from base64 import b64encode

# This reads your video file and displays it in Colab
mp4 = open('my_first_ai_video.mp4','rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML(f"""<video width=600 controls><source src="{data_url}" type="video/mp4"></video>""")
```
